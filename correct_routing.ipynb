    تعداد مقادیر NaN ستون Comments، بسیار بیشتر از مقادیر غیر NaN آن است. بنابراین این ستون را از دیتاست خود حذف کنید.

````python
# drop comments column
df = df.drop('Comments', axis=1)
df
````

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazirmatn" size=3>
مقادیر گم‌شده ستون‌های FuelEconomy و AvgMovingSpeed را با میانگین این ستون‌ها پر کنید. (مقادیر میانگین‌گرفته‌شده‌را تا دو رقم اعشار با تابع round، گرد کنید)
</font>
</p>

````python
# fill NaN values in the AvgMovingSpeed and FuelEconomy
for col in ["AvgMovingSpeed", "FuelEconomy"]:
    df[col] = df[col].fillna(round(df[col].mean(), 2))

df[['AvgMovingSpeed', 'FuelEconomy']]
````

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazirmatn" size=3>
یکی از راه‌های پرکردن مقادیر گم‌شده، پیش‌بینی آن‌ها با استفاده از مدل‌های یادگیری‌ماشین می‌باشد. برای همین منظور، چند سلول کد زیر را اجرا کنید تا مقادیر NaN ستون‌های TotalTime و Toll پیدا شود. هرچند از خودتان انتظار نداشته باشید هم‌اکنون از این کد سر در بیاورید. هنگامی که درسنامه مربوطه در ادامه دوره‌ را خواندید، دوباره به این بخش از درسنامه سر بزنید.    </font>
</p>

````python
automl_reg = AutoML()
x_train, x_test = df.dropna(subset=['TotalTime']), df[df.TotalTime.isna()]
y_train = df[~df.TotalTime.isna()].TotalTime
# removing the columns with null values
x_train = x_train.drop(columns=['TotalTime','Toll'])
x_test = x_test.drop(columns=['TotalTime','Toll'])
# run the model
automl_reg.fit(x_train, y_train, estimator_list=["rf"], task="regression", verbose=False)
y_pred = automl_reg.predict(x_test)
````

````python
df.loc[df.TotalTime.isna(), 'TotalTime'] = y_pred
df.info()
````

````python
automl_clf = AutoML()
x_train, x_test = df.dropna(subset=['Toll']), df[df.Toll.isna()]
y_train = df[~df.Toll.isna()].Toll
# removing the columns with null values
x_train = x_train.drop(columns=['Toll'])
x_test = x_test.drop(columns=['Toll'])
# run the model
automl_clf.fit(x_train, y_train, task="classification", estimator_list=["rf"], verbose=False)
y_pred = automl_clf.predict(x_test)
````

````python
df.loc[df.Toll.isna(), 'Toll'] = y_pred
df.info()
````

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazirmatn" size=3>
همانطور که در نمودار جعبه‌ای MaxSpeed مشاهده می‌شود، داده‌های پرت زیادی در این ستون وجود دارد (حتی داده‌ی ‌منفی برای سرعت!). بنابراین باید این داده‌ها را شناسایی و حذف کرد.
</br>
در قسمت پایین با استفاده از چارک‌ها و محاسبه IQR، داده‌های پرت ستون MaxSpeed را شناسایی کرده و مقادیر پرت این ستون به شکل سورت شده (از کمترین به بیشترین) در لیست outlier_speed بریزید. همچنین مقادیر پرت ستون MaxSpeed  را با میانگین این ستون با دو رقم اعشار با استفاده از تابع round جایگزین کنید . (توجه کنید داده‌های پرت این ستون، نباید تاثیری بر میانگین بگذارند و بدون درنظر گرفتن آن‌ها باید میانگین محاسبه شود.)

</font>
</p>

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazirmatn" size=3>
اگر outlier را پرینت کنید، خروجی شما باید شبیه به زیر باشد:
</br>

```python
[-138.0, -120.0, -32.5, 112.2, 114.4, ...]

```

</font>
</p>

````python
Q1, Q2, Q3 = df.MaxSpeed.quantile([0.25, 0.5, 0.75])
IQR = Q3 - Q1

outlier_df = df[(df['MaxSpeed'].values < Q1-1.5*IQR) | (df['MaxSpeed'].values > Q3+1.5*IQR)]
outlier_df = outlier_df.sort_values(by=['MaxSpeed'])
outlier_speed = outlier_df['MaxSpeed'].tolist() # Keep this for the print output as requested
print(outlier_speed)
````

````python
# replace outliers with avg of MaxSpeed
# Get indices of outliers
outlier_indices = outlier_df.index

# Calculate the mean of MaxSpeed excluding outliers
mean_maxspeed_without_outliers = round(df.drop(outlier_indices)['MaxSpeed'].mean(), 2)

# Replace outliers in df with the calculated mean
df.loc[outlier_indices, 'MaxSpeed'] = mean_maxspeed_without_outliers
````
